## **Explorando la Función Escalón Unitario**
### Autor: ***Felipe Meza-Obando***



### Introducción

La **función escalón unitario** $ u(t) $, también conocida como función Heaviside, es fundamental en análisis de señales y sistemas. Esta función es útil para describir respuestas de sistemas, **activaciones de señales en ciertos instantes de tiempo** y para resolver ecuaciones diferenciales en el análisis de sistemas lineales invariantes en el tiempo (LTI).

**Definición de la función escalón unitario:**

$$
u(t) = \begin{cases}
1 & \text{si } t \geq 0 \\
0 & \text{si } t < 0
\end{cases}
$$

Para el dominio discreto, se representa como $ u[n] $ y tiene el mismo comportamiento. En esta notebook, exploraremos los diferentes casos y variaciones de esta función, incluidas transformaciones en el tiempo y la amplitud.


### Configuración del Entorno

Primero, importamos las bibliotecas necesarias.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-talk')
```

### Definición de la Función Escalón

Definimos la función escalón unitario tanto para el dominio continuo como discreto.

```python
# Función escalón en el dominio continuo
def escalon_continuo(t):
    return np.where(t >= 0, 1, 0)

# Función escalón en el dominio discreto
def escalon_discreto(n):
    return np.where(n >= 0, 1, 0)
```


### Ejemplo 1: Función Escalón Básica en el Dominio Continuo

Graficamos la función escalón en su forma más básica para observar su comportamiento.

```python
# Rango de tiempo para la señal continua
t = np.linspace(-5, 5, 1000)
u_t = escalon_continuo(t)

# Gráfico de la función escalón continua
plt.figure(figsize=(10, 4))
plt.plot(t, u_t, label="Escalón unitario $u(t)$")
plt.title("Función Escalón Unitario $u(t)$")
plt.xlabel("Tiempo $t$")
plt.ylabel("$u(t)$")
plt.grid(True)
plt.xticks(np.arange(n.min(), n.max()+1, 1))
plt.xlim(-5, 5)
plt.legend()
plt.show()
```

En la gráfica se muestra la **función escalón unitario** $ u(t) $ en el dominio continuo. La función toma el valor de $0$ para $ t < 0 $ y cambia a $1$ para $ t \geq 0 $. Esta característica la convierte en una herramienta útil en sistemas y señales para representar activaciones o "encendidos" a partir de un instante específico (en este caso, $ t = 0 $).


### Ejemplo 2: Función Escalón en el Dominio Discreto

En este caso, graficamos la función escalón para el dominio discreto.

```python
# Rango de índices para la señal discreta
n = np.arange(-10, 10)
u_n = escalon_discreto(n)

# Gráfico de la función escalón discreta
plt.figure(figsize=(10, 4))
plt.stem(n, u_n, basefmt=" ", label="Escalón unitario $u[n]$")
plt.title("Función Escalón Unitario en el Dominio Discreto $u[n]$")
plt.xlabel("Índice $n$")
plt.ylabel("$u[n]$")
plt.grid(True)
plt.xticks(np.arange(n.min(), n.max()+1, 1))
plt.xlim(-9, 9)
plt.legend()
plt.show()
```

Esta gráfica muestra la **función escalón unitario en el dominio discreto** $ u[n] $. La función es $0$ para $ n < 0 $ y $1$ para $ n \geq 0 $, lo que representa una activación en el índice $ n = 0 $ y se mantiene en $1$ para todos los valores posteriores.

### Ejemplo 3: Escalón Invertido $ u(-t) $

El caso $ u(-t) $ invierte el escalón, de modo que el salto se produce en el lado negativo del eje.

```python
u_neg_t = escalon_continuo(-t)

# Gráfico del escalón invertido
plt.figure(figsize=(10, 4))
plt.plot(t, u_neg_t, label="Escalón invertido $u(-t)$", color='orange')
plt.title("Función Escalón Invertido $u(-t)$")
plt.xlabel("Tiempo $t$")
plt.ylabel("$u(-t)$")
plt.grid(True)
plt.xticks(np.arange(n.min(), n.max()+1, 1))
plt.xlim(-5, 5)
plt.legend()
plt.show()
```



Esta gráfica muestra la **función escalón invertido** $u(-t)$, donde el salto ocurre en el lado positivo del eje. Aquí, la función es $1$ para $t \leq 0$ y cambia a $0$ para $t > 0$, lo que representa una inversión en el comportamiento del escalón respecto al tiempo.


### Ejemplo 4: Escalón Desplazado $ u(t - t_0) $

Este caso desplaza la función escalón en el tiempo. La función $ u(t - t_0) $ inicia en $ t = t_0 $.

```python
t0 = 2
u_t_shifted = escalon_continuo(t - t0)

# Gráfico del escalón desplazado
plt.figure(figsize=(10, 4))
plt.plot(t, u_t_shifted, label=f"Escalón desplazado $u(t - {t0})$", color='green')
plt.title("Función Escalón Desplazada $u(t - t_0)$")
plt.xlabel("Tiempo $t$")
plt.ylabel("$u(t - t_0)$")
plt.grid(True)
plt.xticks(np.arange(n.min(), n.max()+1, 1))
plt.xlim(-5, 5)
plt.legend()
plt.show()
```

Esta gráfica muestra la **función escalón desplazada** $u(t - t_0)$, donde el salto ocurre en $t = t_0$. En este caso, $t_0 = 2$, lo que significa que la función es $0$ para $t < 2$ y cambia a $1$ para $t \geq 2$. Este desplazamiento permite activar el escalón en un instante específico, diferente de $t = 0$.


### Ejemplo 5: Escalón Desplazado y Escalado $ u(1 - t) $

Este caso es una combinación de desplazamiento e inversión. La función $ u(1 - t) $ empieza en $ t = 1 $ y se activa en el sentido inverso.

```python
u_1_minus_t = escalon_continuo(1 - t)

# Gráfico del escalón desplazado y escalado
plt.figure(figsize=(10, 4))
plt.plot(t, u_1_minus_t, label="Escalón $u(1 - t)$", color='red')
plt.title("Función Escalón Invertida y Desplazada $u(1 - t)$")
plt.xlabel("Tiempo $t$")
plt.ylabel("$u(1 - t)$")
plt.grid(True)
plt.xticks(np.arange(n.min(), n.max()+1, 1))
plt.xlim(-5, 5)
plt.legend()
plt.show()
```

Esta gráfica muestra la **función escalón invertida y desplazada** $u(1 - t)$. Aquí, el escalón se activa en $t = 1$ y es $1$ para $t \leq 1$, cambiando a $0$ para $t > 1$. Esta combinación de inversión y desplazamiento invierte el sentido del escalón y lo desplaza hacia el instante $t = 1$.

### Ejemplo 6: Combinación de Escalones para Formar Pulsos

Podemos combinar funciones escalón para formar pulsos. Por ejemplo, $ u(t) - u(t - T) $ crea un pulso de duración $ T $.

```python
T = 3
u_pulse = escalon_continuo(t) - escalon_continuo(t - T)

# Gráfico del pulso rectangular
plt.figure(figsize=(10, 4))
plt.plot(t, u_pulse, label=f"Pulso rectangular $u(t) - u(t - {T})$", color='brown')
plt.title("Pulso Rectangular con Funciones Escalón")
plt.xlabel("Tiempo $t$")
plt.ylabel("Amplitud")
plt.grid(True)
plt.xticks(np.arange(n.min(), n.max()+1, 1))
plt.xlim(-5, 5)
plt.legend()
plt.show()
```

Esta gráfica muestra un **pulso rectangular** creado utilizando la diferencia de dos funciones escalón: $u(t) - u(t - 3)$. Aquí, el primer escalón $u(t)$ activa el pulso en $t = 0$, mientras que el segundo escalón $u(t - 3)$ lo desactiva en $t = 3$. El resultado es un pulso de amplitud $1$ que permanece activo entre $t = 0$ y $t = 3$, y es $0$ en los demás instantes de tiempo.

### Ejemplo 7: Escalón Discreto con Desplazamiento y Escala

Finalmente, graficamos una combinación en el dominio discreto, donde aplicamos desplazamiento y escala al argumento del escalón

```python
n0 = 2  # Desplazamiento
A = 1.5  # Escala de amplitud
u_n_shifted_scaled = A * escalon_discreto(n - n0)

# Gráfico del escalón discreto desplazado y escalado
plt.figure(figsize=(10, 4))
plt.stem(n, u_n_shifted_scaled, basefmt=" ", label=f"Escalón $A \cdot u[n - {n0}]$")
plt.title("Escalón Unitario en el Dominio Discreto con Desplazamiento y Escala")
plt.xlabel("Índice $n$")
plt.ylabel("$A \cdot u[n - n_0]$")
plt.grid(True)
plt.xticks(np.arange(n.min(), n.max()+1, 1))
plt.xlim(-9, 9)
plt.legend()
plt.show()
```

Esta gráfica muestra un **escalón unitario en el dominio discreto con desplazamiento y escala**, representado por $A \cdot u[n - n_0]$. En este caso, $A = 1.5$ y $n_0 = 2$, lo que desplaza el escalón hacia $n = 2$ y lo escala a una amplitud de $1.5$. La función es $0$ para $n < 2$ y toma el valor de $1.5$ para $n \geq 2$.
